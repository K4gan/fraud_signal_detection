# Fraud signal EDA

A quick look at imbalance, transaction amount, velocity and behavioral distance.

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append(str(Path.cwd().parent))
from train import make_transaction_data

sns.set_theme(style="whitegrid")
df = make_transaction_data(rows=45000, random_state=23)
df.head()
```

```python
fraud_rate = df["is_fraud"].mean()
fig, ax = plt.subplots(figsize=(6, 4))
sns.countplot(data=df, x="is_fraud", ax=ax)
ax.set_title(f"Fraud class balance ({fraud_rate:.2%} fraud)")
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))
sns.histplot(data=df, x="amount", hue="is_fraud", log_scale=True, ax=axes[0])
sns.boxplot(data=df, x="is_fraud", y="velocity_24h", ax=axes[1])
sns.boxplot(data=df, x="is_fraud", y="distance_from_home_km", ax=axes[2])
plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.barplot(data=df, x="payment_channel", y="is_fraud", ax=axes[0])
sns.barplot(data=df, x="country_match", y="is_fraud", ax=axes[1])
axes[0].tick_params(axis="x", rotation=25)
plt.tight_layout()
```
